# Question 1: For the suburb Altona, it is postulated that a typical property sells for 800,000. Use the data at hand to test this assumption. Is the typical property price really 800,000 or has it increased? Use a significance level of 5%.

```python
import pandas as pd
from scipy.stats import ttest_1samp

data = pd.read_csv("/Users/akhilnair/Downloads/property (1).csv")
data
```

```python
# Filtering the data for the suburb Altona
altona_data = data[data['Suburb'] == 'Altona']
altona_data
```

```python
# Removing rows with missing price values
altona_data = altona_data.dropna(subset=['Price'])
# Performing a one-sample t-test to compare the mean property price in Altona to $800,000
# Null hypothesis: The typical property price in Altona is $800,000
# Alternative hypothesis: The typical property price in Altona is not $800,000

population_mean = 800000

# Performing the t-test
t_stat, p_value = ttest_1samp(altona_data['Price'], population_mean)

# Determine if we reject the null hypothesis (Significance Level=alpha)
alpha = 0.05
reject_null = p_value < alpha

t_stat, p_value, reject_null
```

Since the p-value (0.307) is greater than the significance level of 0.05, we fail to reject the null hypothesis. This means there is not enough evidence to conclude that the typical property price in Altona is different from $800,000. Based on this data, we cannot say that the typical property price has increased significantly.

# Question 2:  For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? Consider months from October till March as winter months and rest as summer months. Use a significance level of 5%. 

```python
from scipy.stats import ttest_ind

# Filtering the data for the year 2016
data_2016 = data[data['Date'].str.contains('2016')]

# Converting the Date column to datetime format
data_2016['Date'] = pd.to_datetime(data_2016['Date'], format='%d/%m/%Y')

# Defining summer and winter months
winter_months = [10, 11, 12, 1, 2, 3]
summer_months = [4, 5, 6, 7, 8, 9]

# Splitting the data into winter and summer months
winter_data = data_2016[data_2016['Date'].dt.month.isin(winter_months)]
summer_data = data_2016[data_2016['Date'].dt.month.isin(summer_months)]

# Removing rows with missing price values
winter_data = winter_data.dropna(subset=['Price'])
summer_data = summer_data.dropna(subset=['Price'])

# Performing an independent t-test to compare the mean property prices between winter and summer months
t_stat, p_value = ttest_ind(winter_data['Price'], summer_data['Price'])

# Determine if we reject the null hypothesis (Significance Level=alpha
alpha = 0.05
reject_null = p_value < alpha

t_stat, p_value, reject_null
```

Since the p-value (0.000053) is less than the significance level of 0.05, we reject the null hypothesis. This means there is a statistically significant difference in property prices between the summer months and winter months in the year 2016.

# Question 3: For the suburb of Abbotsford, what is the probability that out of 10 properties sold, 3 will not have car parking? Use the column car in the dataset. Round off your answer to 3 decimal places. 

```python
from scipy.stats import binom

# Load the data
data = pd.read_csv("/Users/akhilnair/Downloads/property (1).csv")

# Filtering the data for the suburb Abbotsford
abbotsford_data = data[data['Suburb'] == 'Abbotsford']

# Calculating the probability of a property not having car parking
no_car_parking_prob = (abbotsford_data['Car'] == 0).mean()

# Calculating the probability that out of 10 properties sold, 3 will not have car parking
n = 10
k = 3
probability = binom.pmf(k, n, no_car_parking_prob)

# Rounding off the result to 3 decimal places
probability_rounded = round(probability, 3)

no_car_parking_prob, probability_rounded
```

The probability that out of 10 properties sold in the suburb of Abbotsford, 3 will not have car parking is approximately 0.26.
